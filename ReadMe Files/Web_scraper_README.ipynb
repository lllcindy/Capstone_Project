# Risk Factor Scraper

Scrape the risk factor section of the 10-K/10-Q reports from edgar database.

---------------------

## Prerequisites

Download/Update the ticker.txt file from SEC website from the following link:<br>
https://www.sec.gov/about/webmaster-faq.htm#developers <br>
make sure to have the ticker.txt file within the directory of the scraper notebook.

## Running the Scraper

First, run all the cells in the "Risk_Factor_Scraper.ipynb". The code will create a folder "risk_factors" (if not already exist) to store the output files.

run the function: 
```
Risk_Factor_Scraper()
```
By default the function will scrape the reports of all the companies for the most recent 3 years.<br>
User can also give the following 2 inputs:<br>
1. ticker_list (eg. [aapl, amzn])<br>
2. year_list (eg.[2018, 2017])<br>

with the input the function will scrape the reports according to the ticker provided within the years in the input year list.

##### Please note that our scraper function currently has some trouble capturing all the patterns for getting the risk factor section due to differences of the report structure among all companies.

## Output

The scraper will have the risk factor section text stored into text files under the folder "risk_factors" for each report with the naming convention of "ticker_filingDate_10K/10Q".
